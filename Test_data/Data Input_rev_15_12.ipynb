```python
metric = temp.quantile(
    q=0.5,                      # The percentile to calculate
    axis=0,                     # The axis to calculate the percentile on
    numeric_only=True,          # To calculate only for numeric columns
    interpolation='linear'      # The type of interpolation to use when the quantile is between 2 values
)
```

```python
metric
```

```python
speed = temp[temp['v_act_kph']> metric[0]]['v_act_kph'].mean()
```

```python
# Initialise the subplot function using number of rows and columns
figure, axis = plt.subplots(2, 2)

# For Lever Force
axis[0, 0].scatter(np.arange(0,len(temp),1),temp['F_act_lever_filtered'])
axis[0, 0].set_title("Lever Force")
  
# For Fluid Pressure
axis[0, 1].scatter(np.arange(0,len(temp),1),temp['P_act_brake_fluid'])
axis[0, 1].set_title("Fluid Pressure")
  
# For Disc Temperature
axis[1, 0].scatter(np.arange(0,len(temp),1),temp['t_act_infrared_filtered'])
axis[1, 0].set_title("Disc Temperature")
  
# For Velocity
axis[1,1].scatter(np.arange(0,len(temp[temp['v_act_kph']> metric[0]]),1),temp[temp['v_act_kph']> metric[0]]['v_act_kph'])
axis[1,1].set_title("Velocity")


l = 2
m = 2

plt.suptitle(f'Lever force = {metric[-1]} N and Velocity = {int(speed)} kmph', fontsize=14)
plt.suptitle(f'Lever force = {50} N and Velocity = {int(speed)} kmph', fontsize=14)
figure.tight_layout()
  
# Combine all the operations and display
plt.show()
```
